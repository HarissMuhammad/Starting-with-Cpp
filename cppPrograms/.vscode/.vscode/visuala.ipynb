```python
import matplotlib.pyplot as plt
```

```python
def collatz_sequence(n):
    sequence = [n]
    while (n != 1):
        if n % 2 == 0:
            n = n/2
        else:
            n = 3*n+1
        sequence.append(n)
    return sequence
```

```python
user_input = input("Enter Any Positive integer")
user_input_list = user_input.split()

Values = []

for i in user_input_list:
    if i.isdigit():
        Values.append(int(i))
if not Values:
    print("Enter a valid value")
else:
    plt.figure(figsize=(10, 6))
    for x in Values:
        collatz_conjecture = collatz_sequence(x)
        plt.plot(collatz_conjecture, marker='o', label=f"Starting from{x}")

    plt.title("Collatz Conjecture")
    plt.xlabel("Step")
    plt.ylabel("Value")
    # plt.legend(loc ="upper right")
    plt.show()
```
