```python
import matplotlib.pyplot as plt
```

```python
def collatz_conjecture(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n/2
        else:
            n = 3*n+1
        sequence.append(n)
    return sequence


user_input_string = input("Enter Positive integers seperated by spaces")
user_input_list = user_input_string.split()

values = []
for val in user_input_list:
    if val.isdigit():
        values.append(int(val))
if not values:
    print("Invalid input")
else:
    plt.figure(figsize=(10, 6))
    for value in values:
        collatz_sequence = collatz_conjecture(value)
        plt.plot(collatz_sequence, marker='o',
                 label=f"starting from value: {value}")
```

```python

plt.title("collatz conjecture")
plt.xlabel("step")
plt.ylabel("Value")
plt.legend(loc='upper right')
plt.show()
```
